# num_sold by date

```python
for i in df.country.unique():
    for j in df.store.unique():
        for k in df['product'].unique():
            plt.figure()
            plt.plot(
                df.loc[(df['country'] == i)&(df['store'] == j)&(df['product']== k), 'date'], 
        df.loc[(df['country'] == i)&(df['store'] == j)&(df['product']== k), 'num_sold'])
            plt.title(f'{i}-{j}-{k}')
            plt.xticks([])
```
